```python
import pandas as pd
```

```python
# create a sample time series data with duplicates

data = pd.DataFrame({'timestamp': ['2023-07-01', '2023-07-02', '2023-07-03', '2023-07-03'], 'value': [1,2,3,3]})
```

```python
data
```

```python
# remove duplicates based on timestamp and values

data = data.drop_duplicates(subset=['timestamp', 'value'], keep='first')

print(data)
```
